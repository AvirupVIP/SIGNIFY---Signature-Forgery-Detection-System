```python
import os
import pandas as pd

ROOT_DIR = "processed_signature\processed"
OUTPUT_CSV = "signature_dataset.csv"

data = []

for split in ["train", "test"]:
    split_path = os.path.join(ROOT_DIR, split)

    for folder in os.listdir(split_path):
        folder_path = os.path.join(split_path, folder)

        if not os.path.isdir(folder_path):
            continue

        # Determine label and person ID
        if "_forg" in folder:
            person_id = int(folder.replace("_forg", ""))
            label = 1   # forged
        else:
            person_id = int(folder)
            label = 0   # genuine

        for file in os.listdir(folder_path):
            if file.lower().endswith((".png", ".jpg", ".jpeg")):
                
                # remove "processed/" from path
                relative_path = os.path.join(split, folder, file)
                relative_path = relative_path.replace("\\", "/")

                data.append({
                    "image_path": relative_path,
                    "person_id": person_id,
                    "label": label,
                    "split": split
                })

df = pd.DataFrame(data)
df = df.sort_values(by=["split", "person_id"])
df.to_csv(OUTPUT_CSV, index=False)

print("CSV created successfully!")
print("Total images:", len(df))
```
