```python
import os
import shutil

#  Change this to your folder path
base_path = r"C:\Users\aviru\OneDrive\Desktop\MyCodes\python\ML\ExtraProcessing2\sample_Signature\forged"

for file in os.listdir(base_path):
    if file.endswith(".png"):
        name_without_ext = file[:-4]  # remove .png
        
        # Take last 3 digits
        last_three = name_without_ext[-3:]
        
        if last_three.isdigit():
            number = int(last_three)
            
            # Add 120
            folder_number = 120 + number
            folder_name = f"{folder_number}_forg"
            
            folder_path = os.path.join(base_path, folder_name)
            os.makedirs(folder_path, exist_ok=True)

            shutil.move(
                os.path.join(base_path, file),
                os.path.join(folder_path, file)
            )

print("✅ Files organized successfully!")
```
