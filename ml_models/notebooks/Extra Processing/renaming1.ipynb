```python
import os

base_path = r"C:\Users\aviru\OneDrive\Desktop\MyCodes\python\ML\Extra Processing\Processed_extra\Fake"  # change this

folders = sorted(os.listdir(base_path))  # get folder names
start_number = 70  # starting from 070

for i, folder in enumerate(folders):
    old_path = os.path.join(base_path, folder)
    
    if os.path.isdir(old_path):  # make sure it's a folder
        new_name = f"{start_number + i:03d}_forg"  # 3-digit format
        new_path = os.path.join(base_path, new_name)
        
        os.rename(old_path, new_path)

print("Renaming done ✅")
```
